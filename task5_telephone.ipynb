```python
def add_contact(contacts):
    name = input("Enter your name: ")
    phone = input("Enter your phone: ")
    email = input("Enter your email: ")
    address = input("Enter your address: ")
    contacts[phone] = {
        "name": name,
        "phone": phone,
        "email": email,
        "address": address
    }
    print("Contact added successfully")
    return contacts

def view_contacts(contacts):
    if not contacts:
        print("No contacts available.")
        return
    for phone, details in contacts.items():
        print(f"Name: {details['name']}, Phone: {phone}, Email: {details['email']}, Address: {details['address']}")

def search_contact(contacts):
    answer = input("Press 1 if you want to search by name and press 2 if you want to search by phone number: ")
    if answer == "1":
        name = input("Enter the name you want to search for: ")
        for phone, details in contacts.items():
            if details["name"].lower() == name.lower():
                print(f"Found contact - Name: {details['name']}, Phone: {phone}, Email: {details['email']}, Address: {details['address']}")
                return
    elif answer == "2":
        phone = input("Enter the phone number you want to search for: ")
        if phone in contacts:
            details = contacts[phone]
            print(f"Found contact - Name: {details['name']}, Phone: {phone}, Email: {details['email']}, Address: {details['address']}")
            return
    print("Contact not found.")

def delete_contact(contacts):
    answer = input("Press 1 if you want to delete by name and press 2 if you want to delete by phone number: ")
    if answer == "1":
        name = input("Enter the name you want to delete: ")
        for phone in list(contacts.keys()):
            if contacts[phone]["name"].lower() == name.lower():
                del contacts[phone]
                print("Contact deleted successfully.")
                return
    elif answer == "2":
        phone = input("Enter the phone number you want to delete: ")
        if phone in contacts:
            del contacts[phone]
            print("Contact deleted successfully.")
            return
    print("Contact not found.")

def update_contact(contacts):
    choice = input("Enter what you want to update:\n1. Name\n2. Phone\n3. Email\n4. Address\n")
    if choice == "1":
        old_name = input("Enter the name you want to update: ")
        for phone, details in contacts.items():
            if details["name"].lower() == old_name.lower():
                new_name = input("Enter the new name: ")
                details["name"] = new_name
                print("Contact updated successfully.")
                return
    elif choice == "2":
        old_phone = input("Enter the phone number you want to update: ")
        if old_phone in contacts:
            new_phone = input("Enter the new phone number: ")
            contacts[new_phone] = contacts.pop(old_phone)
            contacts[new_phone]["phone"] = new_phone
            print("Contact updated successfully.")
            return
    elif choice == "3":
        phone = input("Enter the phone number of the contact whose email you want to update: ")
        if phone in contacts:
            new_email = input("Enter the new email: ")
            contacts[phone]["email"] = new_email
            print("Contact updated successfully.")
            return
    elif choice == "4":
        phone = input("Enter the phone number of the contact whose address you want to update: ")
        if phone in contacts:
            new_address = input("Enter the new address: ")
            contacts[phone]["address"] = new_address
            print("Contact updated successfully.")
            return
    print("Invalid input.")

def choose():
    contacts = {}
    while True:
        print("\nContact Management System")
        print("1. Add Contact")
        print("2. View Contacts")
        print("3. Search Contact")
        print("4. Delete Contact")
        print("5. Update Contact")
        print("6. Exit")
        answer = input("Enter your choice: ")

        if answer == "1":
            add_contact(contacts)
        elif answer == "2":
            view_contacts(contacts)
        elif answer == "3":
            search_contact(contacts)
        elif answer == "4":
            delete_contact(contacts)
        elif answer == "5":
            update_contact(contacts)
        elif answer == "6":
            print("Exit")
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    choose()
```
